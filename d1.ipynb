```python
with open('d1.txt') as f:
    input_data = f.read()
```

```python
bags = input_data.split("\n\n")
```

```python
maxs = []
```

```python
max_cal = 0
for bag in bags:
    ns = bag.split('\n')
    sum_ = sum(int(n) for n in ns if n)
    max_cal = max(max_cal, sum_)
    maxs.append(sum_)
```

```python
maxs.sort()
```
